### Now lets code and see how it works 

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])    #Creating  a single layered neural network
```

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```
model.fit(xs, ys, epochs=500) #Training the model 
```

```
print(model.predict([10.0])) #here we are predicting what could be the outcome for 10
```

```
print(model.predict([11.0]))
```
