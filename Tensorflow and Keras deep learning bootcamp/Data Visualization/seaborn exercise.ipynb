```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diamonds.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x= 'price', y = 'carat', data = df)
```

```python
# as we can see that there are many points on top of each other, its very hard to locate the points 
# so by adding some paramnetes like alpha =1 that means , its totally opaque, and alpha = 0 that means transparent
plt.figure(figsize=(12,3))
sns.scatterplot(x= 'price', y = 'carat', data = df, alpha = 0.1, edgecolor = None)
```

```python
# create hist of price 
plt.figure(figsize=(12,3))
sns.distplot(df['price'], kde = False)
plt.xlim(0,18000)
```

```python
#countplot
sns.countplot(x= 'cut', data = df)
```

```python
#boxplot 
sns.boxplot(x = 'cut', y = 'price', data = df)
```
